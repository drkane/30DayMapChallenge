```python
import folium
```

```python
route = '../02-lines/serum-run-to-nome.geojson'
```

```python
m = folium.Map(
    location=[45.5236, -122.6750],
    tiles='Stamen Terrain'
)
```

```python
line = folium.GeoJson(
    route,
    name='geojson'
)
line.add_to(m)
m.fit_bounds(line.get_bounds())
```

```python
folium.Marker([
    line.data["features"][0]['geometry']['coordinates'][0][1],
    line.data["features"][0]['geometry']['coordinates'][0][0],
], popup=folium.Popup("Nenana", show=True, sticky=True)).add_to(m)
folium.Marker([
    line.data["features"][0]['geometry']['coordinates'][-1][1],
    line.data["features"][0]['geometry']['coordinates'][-1][0],
], popup=folium.Popup("Nome", show=True, sticky=True)).add_to(m)
```
